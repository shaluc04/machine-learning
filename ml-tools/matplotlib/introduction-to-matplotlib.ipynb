## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in jupyter
import matplotlib.pyplot as plt
%matplotlib inline

# 1. Prepare data
x = [1,2,3,4]
y = [1,4,9,16]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot the data
ax.plot(x,y)

# 4. Customize plot
ax.set(title='Simple Plot',
       xlabel='x-axis',
       ylabel='y-axis')

# 5. Save & show (save the whole figure)
fig.savefig('..\..\images\sample-plot.png')
```

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0,10,100)
```

```python
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x**2)  # default in matplotlib is line plot
plt.show()
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x, np.exp(x))
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# Make a plot from dictionary
nut_butter_prices = {'Almond butter':10, 
                     'Peanut Butter':8, 
                     'Cashew Butter':12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values(), width=0.4, color='green')
ax.set(title="Dan's Nut Butter Store",
       xlabel='Butter Types',
       ylabel='Prices ($)')
plt.show()
```

```python
# Horizontal bar plot
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), nut_butter_prices.values(), height=0.5, color='orange')
ax.set(title="Dan's Nut Butter Prices",
       xlabel='Prices')
plt.show()
```

```python
# Make some data
x = np.random.randn(1000)  # generates random samples from normal distribution
```

```python
# Plot the data for histograms and plot it
fig, ax = plt.subplots()
ax.hist(x)   # hist organizes a group of data points into logical ranges/bins. 
plt.show()
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,7))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10), color='green')
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values(), width=0.5, color='red')
ax4.hist(np.random.randn(1000), color='orange')
plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,7))

# Plot to each different axis (index)
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10), color = 'green')
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values(), width=0.5, color='red')
ax[1,1].hist(np.random.randn(1000), color='orange')


plt.show()
```

```python
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
# pyplot method = plotting directly from pandas dataframe
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             figsize=(20,8));
```

```python
# matplotlib's OO (object oriented) method mixed with pyplot
fig, ax = plt.subplots(figsize=(20,8))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
# ax.set_xlim([45,100]);
```

```python
over_50.target.values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

# Customize the plot
ax.set(title='Heart disease and Cholestrol levels',
       xlabel='Age',
       ylabel='Cholestrol');

# add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# add a horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

# add data to ax0
scatter_1 = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])

# Customize ax0
ax0.set(title='Heart disease and cholestrol levels',
        ylabel='Cholestrol')

# add a legend to ax0
ax0.legend(*scatter_1.legend_elements(), title='Target')

# add a horizontal line to ax0
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')


# add data to ax1
scatter_2 = ax1.scatter(x=over_50['age'],
                        y=over_50['thalach'],
                        c=over_50['target'])

# customize ax1
ax1.set(title='Heart disease and max heart rate levels',
        xlabel='Age',
        ylabel='Max heart rate')

# Add legend to ax1
ax1.legend(*scatter_2.legend_elements(), title='Target');

# Add a horizontal line to ax1
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--')

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=14, fontweight='bold');
```

```python
# set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')  # this changes the color scheme

# Customize the plot
ax.set(title='Heart disease and Cholestrol levels',
       xlabel='Age',
       ylabel='Cholestrol');

# add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# add a horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
# Customizing x and y axis limits


# subplot of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

# add data to ax0
scatter_1 = ax0.scatter(x=over_50['age'],
                        y=over_50['chol'],
                        c=over_50['target'],
                        cmap='winter')

# Customize ax0
ax0.set(title='Heart disease and cholestrol levels',
        ylabel='Cholestrol')

# set x axis limit
ax0.set_xlim([50,80])

# add a legend to ax0
ax0.legend(*scatter_1.legend_elements(), title='Target')

# add a horizontal line to ax0
ax0.axhline(y=over_50['chol'].mean(),
            linestyle='--')


# add data to ax1
scatter_2 = ax1.scatter(x=over_50['age'],
                        y=over_50['thalach'],
                        c=over_50['target'],
                        cmap='winter')

# customize ax1
ax1.set(title='Heart disease and max heart rate levels',
        xlabel='Age',
        ylabel='Max heart rate')

# change the x axis limit
# ax1.set_xlim([50,80])

# change ax1 y axis limit
ax1.set_ylim([60,200])

# Add legend to ax1
ax1.legend(*scatter_2.legend_elements(), title='Target');

# Add a horizontal line to ax1
ax1.axhline(y=over_50['thalach'].mean(),
            linestyle='--')

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=14, fontweight='bold');
```
